## Define CSS Styles

### Subtask:
Create the `styles.css` file to style the portfolio with a minimal aesthetic, full-width images, and small sans-serif typography.


**Reasoning**:
I will create a Python script to generate the `style.css` file. This script will define a string containing the CSS rules for a minimal aesthetic, including global resets, typography settings, grid layout for the gallery, and styling for the header and footer, and then write this string to a file named `style.css`.



````python
css_content = """
/* Global Reset */
* {
    margin: 0;
    padding: 0;
    box-sizing: border-box;
}

body {
    font-family: 'Helvetica Neue', Helvetica, Arial, sans-serif;
    font-size: 14px;
    line-height: 1.6;
    color: #333;
    background-color: #fff;
    margin: 0;
}

/* Header */
header {
    padding: 40px 20px;
    text-align: center;
}

header h1 {
    font-weight: normal;
    font-size: 24px;
    letter-spacing: 1px;
    margin-bottom: 20px;
    text-transform: uppercase;
}

nav ul {
    list-style: none;
    padding: 0;
}

nav ul li {
    display: inline-block;
    margin: 0 15px;
}

nav a {
    text-decoration: none;
    color: #666;
    font-size: 12px;
    text-transform: uppercase;
    letter-spacing: 1px;
    transition: color 0.3s ease;
}

nav a:hover {
    color: #000;
}

/* Main Content */
main {
    padding: 20px;
    max-width: 1200px;
    margin: 0 auto;
}

/* Gallery Grid */
.gallery {
    display: grid;
    grid-template-columns: repeat(auto-fill, minmax(300px, 1fr));
    gap: 20px;
}

figure {
    margin: 0;
}

figure img {
    width: 100%;
    display: block;
    height: auto;
    object-fit: cover;
}

figcaption {
    margin-top: 10px;
    font-size: 12px;
    color: #888;
}

/* Footer */
footer {
    text-align: center;
    padding: 40px 20px;
    font-size: 11px;
    color: #aaa;
    margin-top: 40px;
    border-top: 1px solid #eee;
}
"""

with open('style.css', 'w') as f:
    f.write(css_content)

print("style.css created successfully.")
````

## Update CSS with Animations

### Subtask:
Modify `style.css` to add a Material Design-style hover animation to navigation links.


**Reasoning**:
I will append new CSS rules to the `style.css` file to implement a Material Design-style hover animation for the navigation links, as requested.



````python
new_css_rules = """
/* Navigation Hover Animation */
nav a {
    position: relative;
    text-decoration: none;
}

nav a::after {
    content: '';
    position: absolute;
    width: 100%;
    transform: scaleX(0);
    height: 2px;
    bottom: 0;
    left: 0;
    background-color: #0087ca;
    transform-origin: bottom right;
    transition: transform 0.25s ease-out;
}

nav a:hover::after {
    transform: scaleX(1);
    transform-origin: bottom left;
}
"""

with open('style.css', 'a') as f:
    f.write(new_css_rules)

print("style.css updated with hover animations.")
````

**Reasoning**:
I will create a Python script that first converts the HEIC image to JPEG using `pillow-heif` and `PIL`. Then, I will update `index.html` to reference the new image and finally display a preview of the updated site.



````python
from PIL import Image
from pillow_heif import register_heif_opener
from IPython.display import HTML

# 1. Convert HEIC to JPEG
register_heif_opener()

try:
    image = Image.open('/content/IMG_6891.HEIC')
    image.save('project_image.jpg', format='JPEG')
    print("Image converted successfully: project_image.jpg")
except Exception as e:
    print(f"Error converting image: {e}")

# 2. Update index.html
try:
    with open('index.html', 'r') as f:
        html_content = f.read()

    # Replace the first placeholder with the new image
    # The placeholder is: https://via.placeholder.com/300x200?text=Project+1
    target_str = 'https://via.placeholder.com/300x200?text=Project+1'
    if target_str in html_content:
        updated_html = html_content.replace(target_str, 'project_image.jpg')

        with open('index.html', 'w') as f:
            f.write(updated_html)
        print("index.html updated with new image.")
    else:
        print("Placeholder string not found in index.html (or already replaced).")

except Exception as e:
    print(f"Error updating HTML: {e}")
````

## Update Website Content

### Subtask:
Modify index.html to update the caption and style.css to change the gallery layout to a vertical stack.


**Reasoning**:
I will create a Python script to update the `index.html` file by changing the first caption to 'Crab' and append a new CSS rule to `style.css` that changes the gallery layout to a vertical flex column.



````python
# 1. Read index.html
with open('index.html', 'r') as f:
    html_content = f.read()

# 2. Replace the caption
# We only want to replace the first occurrence corresponding to the first project
updated_html = html_content.replace('<figcaption>Project 1</figcaption>', '<figcaption>Crab</figcaption>', 1)

# 3. Write back to index.html
with open('index.html', 'w') as f:
    f.write(updated_html)

# 4. Create new CSS rule
new_css_rule = """
/* Vertical Stack Layout */
.gallery {
    display: flex;
    flex-direction: column;
}
"""

# 5. Append to style.css
with open('style.css', 'a') as f:
    f.write(new_css_rule)

# 6. Print confirmation
print("index.html updated with new caption 'Crab'.")
print("style.css updated with vertical layout rule.")
````

## Generate Embedded Image Preview

### Subtask:
Generate a self-contained HTML preview with embedded images and CSS to verify the layout changes.


**Reasoning**:
I will generate a self-contained HTML preview by embedding the base64-encoded image and the CSS content directly into the HTML structure, allowing for the verification of the latest layout and content changes.



````python
import base64
from IPython.display import HTML

# 1. Read and encode the image
with open('project_image.jpg', 'rb') as img_file:
    b64_string = base64.b64encode(img_file.read()).decode('utf-8')

# 2. Read HTML and CSS
with open('index.html', 'r') as f:
    html_content = f.read()

with open('style.css', 'r') as f:
    css_content = f.read()

# 3. Embed Image as Data URI
# We replace the local file path with the base64 data URI
html_with_image = html_content.replace('src="project_image.jpg"', f'src="data:image/jpeg;base64,{b64_string}"')

# 4. Embed CSS
html_final = html_with_image.replace('<link rel="stylesheet" href="style.css">', f'<style>{css_content}</style>')

# 5. Display
display(HTML(html_final))
````
